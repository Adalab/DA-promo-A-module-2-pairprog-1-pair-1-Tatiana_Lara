```python
import pandas as pd
```

```python
df = pd.read_csv("/mnt/c/Users/USUARIO/Desktop/Adalab/DA-promo-A-module-2-pairprog-1-pair-1-Tatiana_Lara/PANDAS/PANDAS/6_Pandas/union_ataques.csv")
```

```python
df_groupby_paises = pd.DataFrame(df.groupby("Country")["Case Number"].count())
df_groupby_paises.rename(columns = {'Case Number':'Attacks Number'}, inplace = True)
df_groupby_paises_or= df_groupby_paises.sort_values(by='Attacks Number', ascending=False)
df_groupby_paises_or

```

# Ejercicio 2 
¿Cuántos ataques ha habido por año? Devuelve un dataframe. De nuevo ordena los resultados de mayor a menor.

```python
df_groupby_year = pd.DataFrame(df.groupby("Year")["Case Number"].count())
df_groupby_year.rename(columns = {'Case Number':'Attacks Number'}, inplace = True)
df_groupby_year_or= df_groupby_year.sort_values(by='Attacks Number', ascending=False)
df_groupby_year_or
```

# Ejercicio 3
¿Cuántos ataques hubo por año y sexo? Devolved un dataframe solo del año 2000 en adelante.

```python
df_groupby_year_sex = pd.DataFrame(df.groupby(["Year", "Sex "])["Case Number"].count())
df_groupby_year_sex = df_groupby_year_sex.reset_index()
#df_groupby_year_sex = df_groupby_year_sex(["Year"] >2000
df_groupby_year_sex = df_groupby_year_sex[df_groupby_year_sex["Year"] >= 2000]
df_groupby_year_sex 
```

# Ejercicio 4
Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. ¿Cuántos de los ataques fueron letales por sexo? Devuelve un dataframe.

```python
usa_aus_sa =  df[(df["Country"] == "USA") | (df["Country"] == "AUSTRALIA")| (df["Country"] == "SOUTH AFRICA")]
df_groupby_letal_attacks = pd.DataFrame(usa_aus_sa.groupby(["Country", "Sex "])["Fatal (Y/N)"].value_counts())
df_groupby_letal_attacks
#Podriamos filtrar para que nos salgan solo los de yes 
```

# Ejercicio 5
Usando el dataframe creado en el ejercicio anterior. ¿En que pais hubo más ataques?

```python
max_attacks= pd.DataFrame(usa_aus_sa.groupby("Country")["Case Number"].count())
max_attacks_or= max_attacks.sort_values(by='Case Number', ascending=False)
max_attacks_or
```
